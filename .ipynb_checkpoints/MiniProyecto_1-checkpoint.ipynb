# RECONOCIMIENTO DE LEGUAJE DE SEÑAS

#### procedemos a importar las librerias que se pretenden utilizar para procesar las imagenes

```python
# Librerias
import skimage
import skimage.color as skic
import skimage.filters as skif
import skimage.data as skid
import skimage.util as sku
from skimage import util #Manipulacion de imagenes (matrices, formatos)
from scipy.stats import skew, kurtosis #Analisis de datos (asimetria, distribucion)
import matplotlib.pyplot as plt #visualizacion graficos 2D
import numpy as np #Manipulacion de matrices,transformacion de imagenes o señales
import pandas as pd #Manipulacion y analisis de datos estructurados (dataframes)
```

#### Funciones para visualizar la imagen

```python
def showPair(img,filt_img):
    fig, ax = plt.subplots(1, 2, figsize=(10, 10))
    ax[0].imshow(img, cmap=plt.cm.gray)
    ax[0].set_axis_off()
    ax[1].imshow(filt_img, cmap=plt.cm.gray)
    ax[1].set_axis_off()
    plt.show()
```

```python
def show(img):
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.imshow(img, cmap=plt.cm.gray)
    ax.set_axis_off()
    plt.show() 
```

## Filtro Sobel

#### Leemos la imagen

```python
import cv2
imagen = cv2.imread('./images/adel.JPG')
```

#### Para poder visualizar de manera pertinente y manipulable la imagen realizamos una redimencion de ella

```python
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB) 

nuevo_ancho = int(imagen.shape[1] * 0.1)
nuevo_alto = int(imagen.shape[0] * 0.1)
nuevo_tamano = (nuevo_ancho, nuevo_alto)

# Redimensionar la imagen
imagen_redimensionada = cv2.resize(imagen, nuevo_tamano, interpolation=cv2.INTER_AREA) 

# Guardar la imagen redimensionada
cv2.imwrite('./images_redimen/01.jpg', cv2.cvtColor(imagen_redimensionada, cv2.COLOR_RGB2BGR)) 
```

```python
plt.imshow(imagen_redimensionada)
```

#### Siendo las imagenes del dataframe de una resolucion de 4k, para este miniproyecto las redimensionamos a un tamaño mas manipulable y visible en cualquier pantalla que es al 10%, que equivale alrrededor de los 500 px

#### Luego procedemos a identificar el borde de la seña que se pretende utilizar

```python

obimg = skif.sobel(imagen_redimensionada)
showPair(imagen_redimensionada,obimg)
# Verificar los valores mínimo y máximo
print(f"Valor mínimo: {imagen_redimensionada.min()}")
print(f"Valor máximo: {imagen_redimensionada.max()}")
```

#### Como podemos visualizar en la imagen de la derecha los limites de borde de la mano que esta realizando la seña (adelante) es muy precaria, por lo que para mejor identificacion procedemos a realizar el mismo analisis pero con una imagen en escala de grises como se muestra a continuacion

## Normalizar

```python
img = skic.rgb2gray(imagen_redimensionada)

# Convertir la imagen a tipo de dato float32
imagen_float = img.astype(np.float32)

# Normalizar la imagen
imagen_normalizada = cv2.normalize(imagen_float, None, 0, 1, cv2.NORM_MINMAX)

# Convertir la imagen normalizada de nuevo a uint8 para visualización
imagen_normalizada = (imagen_normalizada * 255).astype(np.uint8)

# Verificar los valores mínimo y máximo
print(f"Valor mínimo: {imagen_normalizada.min()}")
print(f"Valor máximo: {imagen_normalizada.max()}")

# Mostrar la imagen normalizada
show(imagen_normalizada) 
```

```python
obimg2 = skif.sobel(img)
showPair(img,obimg2)
```

#### como se visualiza en la anterior figura se puede identificar con mayor presicion el borde de la imagen que se desea; este procedimiento (redimension, manipulacion escala de grises e identificacion de limites de borde) se realiza con las demas imagenes que deseamos implementar mas adelante

```python
import cv2
import os

# Rutas de las imagenes
src_dir = './images/'
dst_dir = './images_redimen/'

# arreglo de imagenes
imgs = [f for f in os.listdir(src_dir) if f.endswith(('.JPG', '.png', '.jpeg'))]

# iterar imagenes
for i, img in enumerate(imgs):
    
    img_path = os.path.join(src_dir, img)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Redimensionar
    h, w, _ = img.shape
    new_size = (int(w * 0.1), int(h * 0.1))
    img = cv2.resize(img, new_size, interpolation=cv2.INTER_AREA)

    # Guardar
    out_path = os.path.join(dst_dir, f'{i+1:02}.jpg')
    cv2.imwrite(out_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

    # escala de grises
    imagen_gris = skic.rgb2gray(imagen_redimensionada)
    
    # filtro Sobel
    imagen_sobel = skif.sobel(imagen_gris)

    # Guardar la imagen de escala de grises y sobel
    ruta_guardar_sobel = os.path.join(dst_dir, f'{i+1:02}_gray_&_sobel.jpg')
    cv2.imwrite(ruta_guardar_sobel, (imagen_sobel * 255).astype('uint8'))  # Convertir a rango adecuado y guardar
```

## Filtrado de Ruido

#### Para la implementacion del proyecto se puede presentar un ruido aditivo al momento querer identificar la seña que el usuario pretende ejecutar por lo que se tiene que generar un filtrado de la imagen para asi poderla utilizar y generar su respectivo comando de movimiento

#### para este ejercicio se implementa un sigma alto para generar una imagen bastante borrosa, desenfocada, detalles y bordes poco finos como se muestra a continuacion

```python
import matplotlib.pyplot as plt
imgLP_Gauss = skif.gaussian(imagen_normalizada, 7)
showPair(imagen_normalizada,imgLP_Gauss)
```

```python
imgHP_AfterSmoothing = img - imgLP_Gauss
showPair(img,imgHP_AfterSmoothing)
```

#### Como puede observase en la ultima imagen se percibe que al restar la imagen suavizada de la original, se realzan los detalles de la seña de forma mas fina junto con sus bordes, generando una imagen que facilita la detección de características relevantes como los contornos, lo que es fundamental para tarea del reconocimiento del gesto
