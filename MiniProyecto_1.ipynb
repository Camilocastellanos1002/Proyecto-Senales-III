# RECONOCIMIENTO DE LEGUAJE DE SEÑAS PARA CONTROLAR AUTOMOVIL EN PANTALLA

procedemos a importar las librerias que se pretenden utilizar para procesar las imagenes

```python
# Librerias
import skimage
import skimage.color as skic
import skimage.filters as skif
import skimage.data as skid
import skimage.util as sku
from skimage import util #Manipulacion de imagenes (matrices, formatos)
from scipy.stats import skew, kurtosis #Analisis de datos (asimetria, distribucion)
import matplotlib.pyplot as plt #visualizacion graficos 2D
import numpy as np #Manipulacion de matrices,transformacion de imagenes o señales
import pandas as pd #Manipulacion y analisis de datos estructurados (dataframes)
```

Leemos la imagen

```python
import cv2
imagen = cv2.imread('./images/adel.JPG')
```

Para poder visualizar la imagen realizamos una redimencion de ella

```python
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB) 

nuevo_ancho = int(imagen.shape[1] * 0.1)
nuevo_alto = int(imagen.shape[0] * 0.1)
nuevo_tamano = (nuevo_ancho, nuevo_alto)

# Redimensionar la imagen
imagen_redimensionada = cv2.resize(imagen, nuevo_tamano, interpolation=cv2.INTER_AREA) 
```

```python
def showPair(img,filt_img):
    fig, ax = plt.subplots(1, 2, figsize=(10, 10))
    ax[0].imshow(img, cmap=plt.cm.gray)
    ax[0].set_axis_off()
    ax[1].imshow(filt_img, cmap=plt.cm.gray)
    ax[1].set_axis_off()
    plt.show()
```

```python
def show(img):
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.imshow(img, cmap=plt.cm.gray)
    ax.set_axis_off()
    plt.show() 
```

```python
plt.imshow(imagen_redimensionada)
```

```python

obimg = skif.sobel(imagen_redimensionada)
showPair(imagen_redimensionada,obimg)
```

```python
img = skic.rgb2gray(imagen_redimensionada)
show(img) 
```

```python
obimg2 = skif.sobel(img)
showPair(img,obimg2)
```

```python
# Guardar la imagen redimensionada
cv2.imwrite('./images_redimen/01.jpg', cv2.cvtColor(imagen_redimensionada, cv2.COLOR_RGB2BGR)) 
```
